```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Company = ['Google','Infosysis','MicroSoft','Amzon']
Revenue= [96,78,56,89]
Profit = [12,32,21,43]
```

```python
ypos = np.arange(len(Company))
ypos
```

```python
plt.bar(ypos,Revenue,color = 'red')
```

```python
plt.xticks(ypos,Company)
plt.bar(ypos,Revenue,color = 'red')
```

```python
plt.xticks(ypos,Company)
plt.ylabel("Revenue")
plt.title("Company Revenue")
plt.bar(ypos,Revenue,color = 'red')
```

```python
plt.xticks(ypos,Company)
plt.ylabel("Revenue")
plt.title("Company Revenue")
plt.bar(ypos,Revenue,color = 'pink',label = 'Revenue')
plt.legend()
```

```python
plt.xticks(ypos,Company)
plt.ylabel("Revenue")
plt.title("Company Revenue")
plt.bar(ypos,Revenue,label = 'Revenue')
plt.bar(ypos,Profit,label='Profit')
plt.legend()
```

```python
plt.xticks(ypos,Company)
plt.ylabel("Revenue")
plt.title("Company Revenue")
plt.barh(ypos,Revenue,label = 'Revenue')
plt.barh(ypos,Profit,label='Profit')
plt.legend()
```

```python
xpos = np.arange(len(Company))
xpos
```

```python
plt.xticks(ypos,Company)
plt.ylabel("Revenue")
plt.title("Company Revenue")
plt.bar(xpos-0.2,Revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,Profit, width=0.4,label="Profit")
plt.legend()
```

```python
plt.xticks(xpos,Company)
plt.ylabel("Revenue")
plt.title("Company Revenue")
plt.barh(xpos,Revenue, label="Revenue")
plt.barh(xpos,Profit,label="Profit")
plt.legend()
```
