```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.2f%%')
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.2f%%',shadow=True,explode=[0,0.1,0.1,0,0])
```

```python
plt.title("My Monthly Expenses")
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.2f%%',shadow=True,explode=[0,0.1,0.1,0,0],startangle=145)
plt.show()
```
