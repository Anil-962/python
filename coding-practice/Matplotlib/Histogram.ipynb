```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [89,90,100,102,112,87,86,97,95,105,90]
plt.hist(blood_sugar,rwidth=.5)
```

```python
plt.hist(blood_sugar,bins=10)
```

```python
plt.hist(blood_sugar,bins=[90,100,110],color='g',histtype='step')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Frequency")
plt.title("Blood Sugar Histogram")
plt.hist(blood_sugar,bins=10,rwidth=.90)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Sugar Level")
plt.ylabel("Frequency")
plt.title("Blood Sugar Histogram")
plt.hist([blood_sugar_men,blood_sugar_women],bins=10,rwidth=.90,color=['green','red'])
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Sugar Level")
plt.ylabel("Frequency")
plt.title("Blood Sugar Histogram")
plt.hist([blood_sugar_men,blood_sugar_women],bins=10,rwidth=.90,color=['green','red'])
plt.legend(['Men','Women'])
```
