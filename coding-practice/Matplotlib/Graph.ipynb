```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [51,56,45,48,45,67,54]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red')
```

```python
plt.plot(x,y,color='red',linestyle='--')
```

```python
plt.xlabel("Weeks")
plt.ylabel("Temp")
plt.title("Temp Graph")
plt.plot(x,y)
```

```python
plt.xlabel("Weeks")
plt.ylabel("Temp")
plt.title("Temp Graph")
plt.plot(x,y,color='green',linewidth=3,linestyle = 'dotted')
```
