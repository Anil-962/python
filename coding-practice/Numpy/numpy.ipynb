```python
import numpy as np
```

```python
a = np.arange(12).reshape(3,4)
a
```

```python
for row in a:
  print(row)
```

```python
for cell in row:
  print(cell)
```

```python
for x in np.nditer(a,order='F'):
  print(x)
```

```python
for x in np.nditer(a,order='F',flags=['external_loop']):
    print(x)
```

```python
for x in np.nditer(a,op_flags=['readwrite']):
  x[...]=x*x
```
