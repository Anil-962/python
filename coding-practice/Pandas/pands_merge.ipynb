```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperature": [21, 14, 35]
})
df1
```

```python
import pandas as pd
df2 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "humidity": [65,68, 75]
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```

```python
df4 = pd.DataFrame({
    "city": ["new york", "chicago", "orlando","Bengaluru"],
    "temperature": [21, 14, 35,38]
})
df4
```

```python
df5 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df5
```

```python
df6 = pd.merge(df1, df2, on="city",how="outer")
df6
```

```python
df7 = pd.merge(df1, df2, on="city",how="inner")
df7
```

```python
df8=pd.merge(df1,df2,on="city",how="left")
df8
```

```python
df9=pd.merge(df1,df2,on="city",how="right")
df9
```
